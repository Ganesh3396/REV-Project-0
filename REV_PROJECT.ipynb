```python
#DataFrame Creation
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
df=pd.read_json("customers.json")
print(df.to_string())
```

# EDA Data Exploration


```python
# Display the first few rows of the dataset
print(df.head(10))
```

```python
#Provides a concise summary of the DataFrame
print(df.info())
```

```python
# Display the last 3 rows of the DataFrame
print(df.tail(10))
```

```python
# Get the shape of the dataset
print(df.shape)
```

```python
# Display the data types of each column
print(df.dtypes)
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Summary statistics for categorical columns
print(df.describe())
```

```python
# Get the column names of the DataFrame
print(df.columns)
```

```python
# Summary statistics for numerical columns
print("\nSummary Statistics (Numerical Columns):")
print(df.describe())
```

```python
# Summary statistics for categorical columns
print("\nSummary Statistics (Categorical Columns):")
print(df.describe(include=['O']))
```

```python
# Value counts for categorical columns
for column in df.select_dtypes(include=['object']).columns:
    print(f"\nValue Counts for {column}:")
    print(df[column].value_counts())
```

```python
# Handling Missing Data
# Drop rows with missing values
df_cleaned = df.dropna()
print("\nDataFrame after dropping missing values:")
print(df_cleaned.to_string())
```

```python
# Fill missing values fill with the mean for numerical columns)
df_filled = df.fillna(df.mean(numeric_only=True))
print("\nDataFrame after filling missing values with mean (for numeric columns):")
print(df_filled.to_string())
```

```python
#Data Visualization

# Histogram of numerical columns
df.hist(bins=10, figsize=(10, 8))
plt.suptitle("Histograms of Numerical Columns")
plt.show()
```

```python
# Boxplot for numerical columns
df.plot(kind='box', subplots=True, layout=(2,3), figsize=(12, 8), title="Boxplots of Numerical Columns")
plt.suptitle("Boxplots of Numerical Columns")
plt.show()
```

```python
# Correlation heatmap
corr_matrix = df.corr(numeric_only=True)
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
def execute_query(query):
    try:
        # Connect to MySQL
        conn = mysql.connector.connect(**conn_params)
        cursor = conn.cursor(dictionary=True)  # Use dictionary cursor for easy access
        cursor.execute(query)
        records = cursor.fetchall()
        df = pd.DataFrame(records)
        return df
    except mysql.connector.Error as err:
        print(f"Error: {err}")
        return None
    finally:
        # Ensure resources are cleaned up
        cursor.close()
        conn.close()
```
